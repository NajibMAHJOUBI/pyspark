## Import libraries

```python
# Python libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set(style="ticks", color_codes=True)
import pandas as pd

# Spark libraries
from pyspark.sql.types import StructType, StructField, DoubleType, StringType
from pyspark.ml.linalg import DenseVector, VectorUDT
from pyspark.sql.functions import udf
from pyspark.ml.clustering import KMeans
```

## Load datset

```python
# We use the iris data set that is a multivariate data set.
# For convenience, the data set has been dowloaded save in the data directory.

# The schema of our data set is defined as follows :
schema = StructType([
    StructField("sepal_length", DoubleType(), True),
    StructField("sepal_width", DoubleType(), True),
    StructField("petal_length", DoubleType(), True),
    StructField("petal_width", DoubleType(), True),
    StructField("species", StringType(), True)])

# data: load of the iris data set
data = spark.read.csv("data/iris.data", header=False, schema=schema)
```

## Preparing data

For each row in the dataset, the four features are aggregated in a vector. This should be done to use the kmeans model in pyspark

```python
## Python function that defined the aggregation in a vector of the four inputs
def create_features(sepal_length, sepal_width, petal_length, petal_width):
    return DenseVector([sepal_length, sepal_width, petal_length, petal_width])

## User defined function: this function is provided to spark to build the feature vectors
udf_create_features = udf(lambda x,y,z,t: create_features(x, y, z, t), VectorUDT())
```

```python
## A new column called "features" is added to the data set contained the aggregate vectors
data = data.withColumn("features", udf_create_features(data["sepal_length"], data["sepal_width"], data["petal_length"], data["petal_width"]))
```

```python
## Preview of the features columns and their corresponding species
data.select("features", "species").show(5)
```

## A quick definition and compute of a kmeans model

```python
## the kmeans clustereing approach is defined by putting the different parameters
## the various parameters are given in the definition of the kmeans class
kmeans = KMeans(featuresCol="features", predictionCol="prediction", k=2, initMode="k-means||", 
                initSteps=10, tol=1e-4, maxIter=20, seed=1234)
```

```python
## compute the kmeans model on the input data set
model = kmeans.fit(data)
```

```python
## transform the input data set
transform = model.transform(data)
```

```python
## Preview of the transform dat set.
## we can note that a column "prediction" has been added to indicated the cluster at which each row belong
transform.show(5)
```

## Elbow approach

The kmeans method is an unsupervised clustering method. That is to say that the optimal number of clustering is not known. It exists different approaches that can be used to determine the number of cluster. The most common one is the elbow method. 

```python
## Compute the sum of squared distances between the input points and their corresponding cluster centers for various 
## number of cluster from 2 to 10
cost = []
number_cluster = range(2, 11)
for k_cluster in number_cluster:
    kmeans = KMeans(featuresCol="features", predictionCol="prediction", k=k_cluster, initMode="k-means||", initSteps=10, tol=1e-4, maxIter=20, seed=1234)
    cost.append(kmeans.fit(data).computeCost(data))
```

```python
## Plot of the cost function vs the number of cluster
fig, ax = plt.subplots();
ax.plot(number_cluster, cost);
ax.set_title("Elbow approach");
ax.set_xlabel("number of cluster");
ax.set_ylabel("cost");
```

From this plot, we can observe that an elbow can be identified for a number of 3 clusters.

## Cluster centroids

One can obtain the center of each cluster from the kmeans model.

```python
kmeans = KMeans(featuresCol="features", predictionCol="prediction", k=3, initMode="k-means||", initSteps=10, tol=1e-4, maxIter=20, seed=1234)
model = kmeans.fit(data)
centers = model.clusterCenters()
```

```python
df_pandas = model.transform(data).toPandas()
data_pandas = data.toPandas()
```

```python
fig, ax = plt.subplots()
species = data.select("species").distinct().rdd.map(lambda x: x["species"]).collect()
colors = ["r", "b", "g"]
for ind,specy in enumerate(species):
    df_pandas[(data_pandas.species == specy)].plot(kind='scatter', x='sepal_length', y='petal_length',
                    color=colors[ind], label=specy, ax=ax);
for center in centers:
    x_center, y_center = center[0], center[2]
    ax.plot(x_center, y_center, "*", ms=20, c="k")  
    
ax.set_title("Clusters with center");
```
