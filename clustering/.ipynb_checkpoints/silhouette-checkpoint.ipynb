## Import libraries

```python
from pyspark.ml.linalg import Vectors
from pyspark.sql.functions import udf, col
from pyspark.sql.types import FloatType
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.ml.clustering import KMeans
from pyspark.ml.linalg import VectorUDT, DenseVector
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import random
import numpy as np
```

## Load dataset 

We consider a 2D dataset of point. After loading the corresponding data set, we create vector features for each point.

```python
create_features = udf(lambda x, y: DenseVector([x, y]), VectorUDT())
```

```python
data = spark.read.parquet("./data/toy_dataset/").withColumn("features", create_features(col("x"), col("y")))
data.show(3)
data_collect = data.rdd.map(lambda p: (p["x"], p["y"])).collect()
```

```python
x, y = map(lambda x: x[0], data_collect), map(lambda x: x[1], data_collect)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1);
ax.scatter(x, y, c="r");
ax.set_title("Initial data");
```

It is clear that the optimal number of cluster is 16.

## Apply of the silhouette method

We simulate the clustering kmeans method for a range of number of culster. For each number of cluster, 5 seeds number have tested to obtain a smooth plot of the silhouette score according to the number of score.

```python
evaluator = ClusteringEvaluator(predictionCol="prediction", featuresCol="features", metricName='silhouette')
nb_cluster = range(2, 30)
seeds = [random.randint(0, sys.maxint) for _ in range(5)]
silhouette_score = {nb: [] for nb in nb_cluster}
for seed in seeds:
    for nb in range(2, 30):
        kmeans = KMeans(featuresCol="features", predictionCol="prediction", k=nb, initMode="k-means||", 
                    initSteps=10, tol=1e-4, maxIter=20, seed=seed)
        prediction = kmeans.fit(data).transform(data)
        silhouette_score[nb].append(evaluator.evaluate(prediction))
```

```python
silhouette_mean = {key: np.mean(silhouette_score[key]) for key in silhouette_score.keys()}
items = list(silhouette_mean.iteritems())
items.sort(key=lambda p: [0])
nb_cluster, score = map(lambda p: p[0], items), map(lambda p: p[1], items)
fig, ax = plt.subplots(nrows=1, ncols=1);
ax.plot(nb_cluster, score);
ax.set_xlabel("Number of cluster");
ax.set_ylabel("Silhouette score");
```

```python
print("The maximum silhouette method {0} is obtained for {1} clusters.".format(max(score), nb_cluster[score.index(max(score))]))
```
